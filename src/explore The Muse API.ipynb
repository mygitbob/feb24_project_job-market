```python
import requests
import pandas as pd

# Variables for the Adzuna API credentials

api_key = "282a5210653ba97032332e9db4b9408627a9bb511f47b1ece3a24706318e49e2"

# Initialize an empty DataFrame
df_raw = pd.DataFrame()

# Number of pages you want to fetch
pages = 2

for page in range(1, pages + 1):
    url = f"https://www.themuse.com/api/public/jobs?location=Dublin%2C%20Ireland&page={page}&descending=false&api_key={api_key}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        df_page = pd.DataFrame(data['results'])
        
        # Concatenate the current page's DataFrame with the main DataFrame
        df_raw = pd.concat([df_raw, df_page], ignore_index=True)
    else:
        print(f"Failed to fetch data for page {page}. Status code: {response.status_code}")

display(df_raw)
```

```python
# Regular expression pattern with capture groups for extracting the salary range
pattern = r"\$(\d{1,3}(?:,\d{3})*(?:\.\d{2})?)\s*-\s*\$(\d{1,3}(?:,\d{3})*(?:\.\d{2})?)"

extracted_ranges = df_raw['contents'].str.extract(pattern)

df_raw['salary_min'] = extracted_ranges[0]
df_raw['salary_max'] = extracted_ranges[1]

print(df_raw[['salary_min', 'salary_max']])
```

```python
import pandas as pd
import json

file_path_processed = "C:\\Users\\btsonkov\\Git Repo\\feb24_project_job-market\\data\\processed\\muse.com\\muse_proc.merged.2024-03-18 12_20_00.json"

with open(file_path_processed, "r", encoding="utf-8") as file:
    loaded_json_prced = json.load(file)
    
df_muse = pd.DataFrame(loaded_json_prced)

df_muse.head()   
```

```python
# Parse the value of 'name' from the JSON structure into a new column
df_muse['category'] = df_muse['job_categories'].apply(lambda x: x[0]['name'] if x else None)
df_muse['extracted_location'] = df_muse['location'].apply(lambda x: x[0]['name'] if x else None)
```

```python
from geopy.geocoders import Nominatim

def get_country_code(location_string):
    geolocator = Nominatim(user_agent="feb24_project_job-market")
    location = geolocator.geocode(location_string)
    if location:        
        return location.raw['display_name'].split(', ')[-1].upper()
    else:
        return None
```

```python
df_muse['country'] = df_muse['extracted_location'].apply(lambda x: get_country_code(x))
df_muse.head(10) 
```
