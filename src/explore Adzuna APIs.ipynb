```python
import requests
import pandas as pd

# Variables for the Adzuna API credentials
app_id = "bccb7a79"
app_key = "cf5a43a48e02fd703aff0f3a5b20c21f"

# Initialize an empty DataFrame
df_raw = pd.DataFrame()

# Number of pages you want to fetch
pages = 5

for page in range(1, pages + 1):
    url = f"https://api.adzuna.com/v1/api/jobs/gb/search/{page}?app_id={app_id}&app_key={app_key}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()['results']
        
        # Preprocess each job to replace the company object with just the display_name
        for job in data:
            if 'company' in job:
                job['company'] = job['company']['display_name']
            if 'location' in job:
                job['location_area'] = job['location']['area']
                job['location_name'] = job['location']['display_name']
            if 'category' in job:
                job['category_tag'] = job['category']['tag']
                job['category_label'] = job['category']['label']
        
        # Create a DataFrame from the preprocessed data
        df_page = pd.DataFrame(data)
        
        # Concatenate the current page's DataFrame with the main DataFrame
        df_raw = pd.concat([df_raw, df_page], ignore_index=True)
    else:
        print(f"Failed to fetch data for page {page}. Status code: {response.status_code}")

```

```python
df_raw = df_raw.drop(['adref', '__CLASS__', 'longitude', 'latitude', 'location', 'category'], axis=1)

display(df_raw)
```

```python
df_raw['extracted_location'] = df_raw['location_area'].apply(lambda x: x[0] if x else None)
```
